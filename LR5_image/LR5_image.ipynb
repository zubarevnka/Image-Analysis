```python
import numpy as np
from skimage.io import imread, imshow
from skimage.morphology import (closing, opening, binary_closing, binary_opening,binary_erosion, binary_dilation, erosion, dilation, diamond, disk, rectangle, square, star)
from skimage.color import rgb2gray, rgb2yuv, yuv2rgb
import matplotlib.pyplot as plt
from skimage import data
from sklearn.datasets import load_digits
from skimage import img_as_float, img_as_ubyte
from skimage.filters import rank
%matplotlib inline
```

```python
def show_img2(img1, img2, label1='', label2=''):
    fig = plt.figure(figsize=(15, 10))  
    plt.subplot(1,2,1)
    plt.title(label1)
    plt.imshow(img1)
    plt.subplot(1,2,2)
    plt.title(label2)
    plt.imshow(img2)
```

```python
def show_img3(img1, img2, img3, label1='', label2='', label3=''):
    plt.figure(figsize=(15,10))
    plt.subplot(1,3,1)
    plt.title(label1)
    plt.imshow(img1)
    plt.subplot(1,3,2)
    plt.title(label2)
    plt.imshow(img2)
    plt.subplot(1,3,3)
    plt.title(label3)
    plt.imshow(img3)         
    
def show_img4(img1, img2, img3, img4, label1='', label2='', label3='', label4=''):
    plt.figure(figsize=(15,10))
    plt.subplot(2,2,1)
    plt.title(label1)
    plt.imshow(img1)
    plt.subplot(2,2,2)
    plt.title(label2)
    plt.imshow(img2)
    plt.subplot(2,2,3)
    plt.title(label3)
    plt.imshow(img3)
    plt.subplot(2,2,4)
    plt.title(label4)
    plt.imshow(img4)  
```

## Определение возраста дерева по спилу

```python
img_tree = imread('11.jpg')
plt.figure(figsize=(8,8))
img_tree = rgb2gray(img_tree)
imshow(img_tree)
```

```python
img_tree_cut = img_tree[1610:1650, 285:1490]
plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut)
```

```python
from skimage.filters.rank import equalize
img_tree_cut= equalize(img_tree_cut, selem=disk(30))
plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut)
```

```python
img_tree_cut = img_as_float(img_tree_cut>170)
plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut)
```

```python
img_tree_cut = dilation(img_tree_cut, square(2))
plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut)
```

```python
line = np.mean(img_tree_cut, axis = 0)
img_tree_cut = np.array([line]*40)
plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut)
```

```python
img_tree_cut = img_as_float(binary_erosion(img_tree_cut, rectangle(2,2)))
plt.figure(figsize=(20,20))
plt.imshow(img_tree_cut)
```

```python
def get_tree_age(X): 
    count_series = 0
    
    for i in range(len(X)):
        if X[i]>0.3:
            X[i]=1
        else:
            X[i]=0
            
    for i in range(len(X)-1):
        if X[i+1]!=X[i]:
            if X[i]==0:
                count_series+=1 #количество серий
    print('Возраст дерева:', count_series)
```
